```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import datetime as dt
```

```python
plt.style.use('ggplot') 
font = {'family' : 'meiryo'}
matplotlib.rc('font', **font)
```

```python
url = 'https://vincentarelbundock.github.io/Rdatasets/csv/robustbase/ambientNOxCH.csv'
df_sample = pd.read_csv(url, parse_dates=True, index_col=1)

print(type(df_sample))
df = df_sample.iloc[:, 1:]
df.head()
```

```python
df_sample.head()
```

```python
df_monthly = df.copy()
df_monthly.index = df_monthly.index.map(lambda x:x.month)
df_monthly = df_monthly.groupby(level=0).sum()
df_monthly.head()
```

```python
df_monthly.plot.bar(y=["ad", "ba", "se"], alpha=0.6, figsize=(12,3))
```

```python
df_monthly.plot.bar(alpha=0.6, figsize=(12,3))
```

```python
df_monthly.plot.bar(y=['ad', 'ba', 'se'], alpha=0.6, figsize=(12,3), stacked=True, cmap='Blues')
```
