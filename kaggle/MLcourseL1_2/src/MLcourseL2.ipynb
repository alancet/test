```python
import pandas as pd

melb_data = pd.read_csv('../input/melbourne-housing-snapshot/melb_data.csv')

melb_data

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

melb_target = melb_data.Price
melb_target.head()
```

```python
melb_predictors = melb_data.drop(['Price'], axis=1)
melb_predictors.head()
```

```python
melb_numeric_preditors = melb_predictors.select_dtypes(exclude=['object'])
melb_numeric_preditors.head()
```

```python
from sklearn.preprocessing import Imputer
my_imputer = Imputer()
data_with_imputed_values = my_imputer.fit_transform(melb_numeric_preditors)
data_with_imputed_values
```

```python
X_train, X_test, y_train, y_test = train_test_split(melb_numeric_preditors, melb_target, test_size=0.33, random_state=0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
cols_with_missing = [col for col in X_train.columns
                    if X_train[col].isnull().any()]
cols_with_missing
```

```python
reduced_X_train = X_train.drop(cols_with_missing, axis=1)
reduced_X_test = X_test.drop(cols_with_missing, axis=1)
reduced_X_train.head()
```

```python
reduced_X_test.head()
```

```python
def score_dataset(reduced_X_train, reduced_X_test, y_train, y_test):
    rfr_model = RandomForestRegressor()
    #print(reduced_X_train.shape, y_train.shape)
    rfr_model.fit(reduced_X_train, y_train)
    y_pred = rfr_model.predict(reduced_X_test)
    mae = mean_absolute_error(y_test, y_pred)
    return mae
```

```python
print(reduced_X_train.shape, reduced_X_test.shape, y_train.shape, y_test.shape)
print(score_dataset(reduced_X_train, reduced_X_test, y_train, y_test))
```

```python
from sklearn.preprocessing import Imputer

my_imputer = Imputer()
imputed_X_train = my_imputer.fit_transform(X_train)
imputed_X_test = my_imputer.fit_transform(X_test)
print(score_dataset(imputed_X_train, imputed_X_test, y_train, y_test))
```
