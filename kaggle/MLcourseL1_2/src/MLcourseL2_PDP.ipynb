```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer

data = pd.read_csv('../input/melbourne-housing-market/Melbourne_housing_FULL.csv')
data.dropna(axis=0, inplace=True)
y = data.Price

X = data.drop(['Price'], axis=1).select_dtypes(exclude=['object'])
```

```python
X.dropna(axis=0).shape
```

```python
from sklearn.ensemble.partial_dependence import partial_dependence, plot_partial_dependence
from sklearn.ensemble import GradientBoostingRegressor

my_model = GradientBoostingRegressor()

my_model.fit(X, y)

fig, axs = plot_partial_dependence(my_model, features=[0, 1, 2],
                                  X=X, 
                                  feature_names=['Distance', 'Landsize', 'BuildingArea'],
                                  grid_resolution=10)
fig
```
