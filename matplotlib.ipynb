```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y, s=600, c='pink', alpha=0.5, linewidths='2', edgecolors='red')
```

```python
plt.scatter(x, y, marker='*')
```

```python
plt.scatter(x, y)
plt.title('This is a title')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
```

```python
value = np.random.rand(100)

plt.scatter(x, y, c=value, cmap='Blues')
plt.colorbar()
```

```python
plt.scatter(x, y, c=value, cmap='Blues', vmin=0.4, vmax=0.6)
plt.colorbar()
```
